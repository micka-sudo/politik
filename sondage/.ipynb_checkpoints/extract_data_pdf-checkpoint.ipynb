# Extract data to pdf

## Import Librairies



```python
import pandas as pd
import pdfplumber
```

## Load pdf

```python
path = "./pdf/novembre_2021/9241-p-hi-challenges-v22.pdf"
```

```python
pdf = pdfplumber.open(path)

for page in pdf.pages:
    for table in enumerate(page.extract_tables()):
        df = pd.DataFrame(table)
        print(pd.DataFrame(table))
```
